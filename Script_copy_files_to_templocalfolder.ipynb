### Copy files from intermediate storage to another local folder to avoid copying writing files into next arbitrage detection spark

```python
import os
import shutil
import time

# Paths for the temporary and final storage directories
temp_folder_path = "/Users/priyasuresh/Documents/Documents_Priya’s MacBook Pro/SPRING2024/Data 228-Big Data/project/intermediate_storage"
final_folder_path = "/Users/priyasuresh/Documents/Documents_Priya’s MacBook Pro/SPRING2024/Data 228-Big Data/project/temp_folder"

def safely_move_file(src_file, dest_dir):
    """Move files to the destination directory after writing is complete."""
    try:
        # Check if destination directory exists, create if not
        if not os.path.exists(dest_dir):
            os.makedirs(dest_dir)

        # Check if the file is older than 10 seconds to assume writing is done
        if time.time() - os.path.getmtime(src_file) > 10:
            # Move the file
            shutil.move(src_file, os.path.join(dest_dir, os.path.basename(src_file)))
            print(f"File {os.path.basename(src_file)} moved successfully to {dest_dir}")
        else:
            print(f"File {os.path.basename(src_file)} is still being written to.")
    except Exception as e:
        print(f"Error moving file {src_file} to {dest_dir}: {str(e)}")

while True:
    # Iterate over all files in the temporary directory
    for filename in os.listdir(temp_folder_path):
        file_path = os.path.join(temp_folder_path, filename)
        if os.path.isfile(file_path):
            safely_move_file(file_path, final_folder_path)
    
    # Sleep for a short period to avoid high CPU usage
    time.sleep(30)  # Sleep time in seconds, adjust as needed
```
